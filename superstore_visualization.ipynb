```python
#Import all three required libraries
import matplotlib.pyplot as plt 
import seaborn as sns 
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('Profit_every_month.csv')
```

```python
df.iloc[0:1]
```

```python
df.dtypes
```

```python
df=df.astype({'year_month':str})
```

```python
df.dtypes
```

```python
sns.set(rc={'figure.figsize':(12, 5)})

sns.set_style('darkgrid')#This only adds a plot style, making the bottom darker
sns.set_palette('Set2')#This only controls the color palette, makes it nicer.


sns.scatterplot(data=df, x='year_month', y='profit' 
                #,hue='quantity' #controls dot color
                #,alpha=0.8 #controls the dot transparency, 0.5 is half transparent
                #,style='species'#this helps us to discern another categorical variable (species) as dot shape
                ,size='quantity' #adds a numerical variable as size of dots
                ,sizes=(10,200) #adds the dot size ranges
            )



plt.title('Month vs Profit')#adding a title
plt.xlabel('Month') #x title
plt.ylabel('Profit') #y title
plt.xticks(rotation=90);
```

### Following code is for the 'profit_each_year' plot

```python
df=pd.read_csv('Profit_each_year.csv')
```

```python
df.head()
```

```python
sns.set(rc={'figure.figsize':(12, 5)})

sns.set_style('darkgrid')#This only adds a plot style, making the bottom darker
sns.set_palette('Set2')#This only controls the color palette, makes it nicer.


sns.scatterplot(data=df, x='year', y='profit' 
                #,hue='quantity' #controls dot color
                #,alpha=0.8 #controls the dot transparency, 0.5 is half transparent
                #,style='species'#this helps us to discern another categorical variable (species) as dot shape
                ,size='quantity' #adds a numerical variable as size of dots
                ,sizes=(10,200) #adds the dot size ranges
            )



plt.title('year vs profit')#adding a title
plt.xlabel('year') #x title
plt.ylabel('Profit') #y title
plt.xticks(np.arange(2014, 2018, 1),rotation=45);
```

## Bar Plot

```python
df=pd.read_csv('top_ten_customers.csv')
```

```python
df.head()
```

```python
sns.set(rc={'figure.figsize':(12, 5)})

sns.set_style('darkgrid')#This only adds a plot style, making the bottom darker
sns.set_palette('Set2')#This only controls the color palette, makes it nicer.


sns.barplot(data=df, x='customer_name', y='profit' 
                ,hue='segment' #controls bar color
            )

plt.title('Top ten customers and their segment')#adding a title
plt.xlabel('customer name') #x title
plt.ylabel('Profit') #y title
plt.xticks(rotation=45);
```

```python
df=pd.read_csv('region_profit_pie.csv')
```
